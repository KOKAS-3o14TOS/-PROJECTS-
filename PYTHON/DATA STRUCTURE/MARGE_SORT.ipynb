# Merge Sort 
Merge Sort Data Structure and Algorithms Tutorials. (2013, March 15). GeeksforGeeks; GeeksforGeeks. https://www.geeksforgeeks.org/merge-sort/

‌Marge sort is a recursive algorithm that continuously split the array in half until it  cannot be further divided.

1. Initially divide the array into two equal halves:

‌<center><img src="IMG/Merge_sort.png"></center> 

2. These subarrays are further divided into two halves. Now they become array of unit length that can no longer be divided and array of unit length are always sorted.

‌<center><img src="IMG/Merge_sort2.png"></center> 

3. These sorted subarrays are merged together, and we get bigger sorted subarrays.

‌<center><img src="IMG/Merge_sort3.png"></center> 

4. This merging process is continued until the sorted array is built from the smaller subarrays.

‌<center><img src="IMG/Merge_sort4.png"></center> 



```python


# Python program for implementation of MergeSort
 
 
def mergeSort(arr):
    if len(arr) > 1:
 
         # Finding the mid of the array
        mid = len(arr)//2
 
        # Dividing the array elements
        L = arr[:mid]
 
        # Into 2 halves
        R = arr[mid:]
 
        # Sorting the first half
        mergeSort(L)
 
        # Sorting the second half
        mergeSort(R)
 
        i = j = k = 0
 
        # Copy data to temp arrays L[] and R[]
        while i < len(L) and j < len(R):
            if L[i] <= R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
 
        # Checking if any element was left
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
 
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1
 
 
# Code to print the list
def printList(arr):
    for i in range(len(arr)):
        print(arr[i], end=" ")
    print()
 
 
# Driver Code
if __name__ == '__main__':
    arr = [12, 11, 13, 5, 6, 7]
    print("Given array is")
    printList(arr)
    mergeSort(arr)
    print("\nSorted array is ")
    printList(arr)
 
# This code is contributed by Mayank Khanna
```
